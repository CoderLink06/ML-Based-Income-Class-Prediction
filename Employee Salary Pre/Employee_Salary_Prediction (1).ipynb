```python
# Load your library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split
import joblib

# Load the data
data = pd.read_csv("/content/adult 3.csv")

# Handle '?' values in 'occupation' and 'workclass'
# data.occupation.replace({'?': 'Others'}, inplace=True)
# data.workclass.replace({'?': 'NotListed'}, inplace=True)
data.loc[:, 'occupation'] = data['occupation'].replace({'?': 'Others'})
data.loc[:, 'workclass'] = data['workclass'].replace({'?': 'NotListed'})


# Remove specific workclass values
data = data[data['workclass'] != 'Without-pay']
data = data[data['workclass'] != 'Never-worked']

# Remove specific education values
data = data[data['education'] != '5th-6th']
data = data[data['education'] != '1st-4th']
data = data[data['education'] != 'Preschool']

# Drop the 'education' column
data.drop(columns=['education'], inplace=True)

# Filter based on age to remove outliers
data = data[(data['age'] <= 75) & (data['age'] >= 17)]

# Label Encoding for categorical features
categorical_cols = ['workclass', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'native-country']
fitted_encoders = {} # Dictionary to store fitted encoders

for col in categorical_cols:
    encoder = LabelEncoder()
    data[col] = encoder.fit_transform(data[col])
    fitted_encoders[col] = encoder # Save the fitted encoder


# Separate features (x) and target (y)
x = data.drop(columns=['income'])  # input
y = data['income']  # output


# Scale numerical data
scaler = MinMaxScaler()
x = scaler.fit_transform(x)

# Split data into training and testing sets
# Ensure no NaNs before splitting, although previous steps should handle '?'
x_cleaned = x[~np.isnan(x).any(axis=1)]
y_cleaned = y[~y.isna()]

xtrain, xtest, y_train, y_test = train_test_split(x_cleaned, y_cleaned, test_size=0.2, random_state=23, stratify=y_cleaned)

# Save the test data for the Streamlit app
joblib.dump(xtest, 'xtest.pkl')
joblib.dump(y_test, 'y_test.pkl')


# Train and save the models
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier

# Train KNN model
knn = KNeighborsClassifier()
knn.fit(xtrain, y_train)
joblib.dump(knn, 'knn_model.pkl')

# Train Logistic Regression model
lr = LogisticRegression()
lr.fit(xtrain, y_train)
joblib.dump(lr, 'lr_model.pkl')

# Train MLPClassifier model
# Using the parameters from the previous successful run in the notebook
mlp_model = MLPClassifier(solver='adam', hidden_layer_sizes=(5, 2), random_state=2, max_iter=2000)
mlp_model.fit(xtrain, y_train)
joblib.dump(mlp_model, 'mlp_model.pkl')

# Save the scaler
joblib.dump(scaler, 'scaler.pkl')

# Save the fitted encoders (optional but recommended for robust app)
joblib.dump(fitted_encoders, 'fitted_encoders.pkl')


print("Data loading, preprocessing, model training, and saving complete.")
```
